## Example: Extracting spectrogram

````python
import requests
import torchaudio
import matplotlib.pyplot as plt
import torchaudio.functional as F
import torchaudio.transforms as T
import librosa

target_sample_rate = 16000
n_fft = 512

# a function to plot spectrogram
def plot_spectrogram(specgram, title=None, ylabel="freq_bin"):
    fig, axs = plt.subplots(1, 1)
    axs.set_title(title or "Spectrogram (db)")
    axs.set_ylabel(ylabel)
    axs.set_xlabel("frame")
    im = axs.imshow(librosa.power_to_db(specgram), origin="lower", aspect="auto")
    fig.colorbar(im, ax=axs)
    plt.show(block=False)

# url to find the audio sample
url = "https://drwuz.com/CSC3160/materials/audio/LJ001-0048.wav"
# request the url
r = requests.get(url)

# filename, where to save the audio
filename = "LJ001-0048.wav"
with open(filename, 'wb') as f:
    f.write(r.content)

# load audio sample to memory
waveform, sample_rate = torchaudio.load(filename)

# create an object that resamples a waveform
resampler = T.Resample(sample_rate, target_sample_rate, dtype=waveform.dtype)
# resample a waveform from its original sampling rate to the target sampling rate
resampled_waveform = resampler(waveform)

# create an object that extract power spectrogram
spec_trans = T.Spectrogram(n_fft=n_fft, power=2)

# extract spectrogram
spec = spec_trans(resampled_waveform)

print("Shape of waveform: {}".format(waveform.size()))
print("Sample rate of waveform: {}".format(sample_rate))

print("Shape of resampled_waveform: {}".format(resampled_waveform.size()))

plt.figure()
plt.plot(waveform.t().numpy())
plt.title("Waveform Frequency")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plot_spectrogram(spec[0])
````

## (1) [20 marks] Extract mel-spectrogram

In this part, you need to extract power spectorgram (FFT 1024) with 5ms, 10ms, 15ms frameshift, with the same window size 30ms. The speech signal is at 16 kHz sampling rate.
* Visualize each of the power spectorgram (FFT 1024)
* Plot the spectrum of the 150th frame of the 512-dimensional spectrogram with 10 ms frameshift



1. Visualize each of the power spectorgram (FFT 1024)

````python
# Constants
target_sample_rate = 16000
n_fft = 1024
win_length = int(0.03 * target_sample_rate)  # 30ms window size
plt.figure()
plt.plot(waveform.t().numpy())
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Waveform Frequency")


# Frame shifts = 5ms
hop_length_5ms = int(0.005 * target_sample_rate)
spec_trans_5ms = T.Spectrogram(n_fft=n_fft, power=2, hop_length=hop_length_5ms, win_length=win_length)
spec_5ms = spec_trans_5ms(resampled_waveform)
plot_spectrogram(spec_5ms[0], title=f"Power Spectrogram with 5ms samples frame shift")

# Frame shifts = 10ms
hop_length_10ms = int(0.01 * target_sample_rate)
spec_trans_10ms = T.Spectrogram(n_fft=n_fft, power=2, hop_length=hop_length_10ms, win_length=win_length)
spec_10ms = spec_trans_10ms(resampled_waveform)
plot_spectrogram(spec_10ms[0], title=f"Power Spectrogram with 10ms samples frame shift")

# Frame shifts = 15ms
hop_length_15ms = int(0.015 * target_sample_rate)
spec_trans_15ms = T.Spectrogram(n_fft=n_fft, power=2, hop_length=hop_length_15ms, win_length=win_length)
spec_15ms = spec_trans_15ms(resampled_waveform)
plot_spectrogram(spec_15ms[0], title=f"Power Spectrogram with 15ms samples frame shift")
````

2. Plot the spectrum of the 150th frame of the 512-dimensional spectrogram with 10 ms frameshift

````python
frame_150_spectrum = spec_10ms[0, :, 149]  # 150th frame (index 149)
plt.figure()
plt.title("Spectrum of the 150th frame (10ms frame shift)")
plt.xlabel("Frequency Bin")
plt.ylabel("Magnitude")
plt.plot(frame_150_spectrum.numpy())
plt.show()
````
